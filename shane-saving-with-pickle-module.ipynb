# Prep conspiracy data document-term matrix

```python
import pickle
from cytoolz import identity, take
from sklearn.feature_extraction.text import CountVectorizer, ENGLISH_STOP_WORDS
from gensim import corpora, models
from gensim.matutils import Sparse2Corpus
```

```python
def read_corpus(name):
    for line in open(name):
        yield [tok for tok in line.lower().split() if tok.isalpha() and tok not in ENGLISH_STOP_WORDS]
```

```python
V = CountVectorizer(analyzer=identity, min_df=50, max_df=0.50)
X = V.fit_transform(read_corpus('/home/malouf/ling583_s18/conspiracy-100k.txt'))
```

```python
X.shape
```

```python
corpus = Sparse2Corpus(X, documents_columns=False)
```

```python
vocab = corpora.dictionary.Dictionary()
vocab.token2id = dict((w,i) for (i,w) in enumerate(V.get_feature_names()))
vocab.id2token = dict(enumerate(V.get_feature_names()))
```

```python
with  open('conspiracy.dat', 'wb') as f:
    pickle.dump((vocab, corpus), f, -1)
```
