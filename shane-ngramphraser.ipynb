# Phrase extraction

Find common phrases using Mikolov et al.'s (2013) method


```python
import pandas as pd
import spacy
from cytoolz import *
from gensim.models.phrases import Phrases, Phraser
```

```python
nlp = spacy.load('en', disable=['tagger', 'ner', 'parser'])
```

## Load and tokenize data

```python
df = pd.read_csv("../input/phraserdata/phraser_data.csv")
```

```python
def tokenize(text):
    return [tok.orth_ for tok in nlp.tokenizer(text)]
df['tokens'] = df['review_text'].apply(tokenize)
```

## Find bigram phrases

```python
P1 = Phraser(Phrases(df['tokens'], threshold=20))
```

```python
for s in take(5, df['tokens']):
    print(' '.join(P1[s]))
    print()
```

## Find (up to) trigram phrases

```python
P2 = Phraser(Phrases(P1[list(df['tokens'])], threshold=15))
```

```python
for s in take(5, df['tokens']):
    print(' '.join(P2[P1[s]]))
    print()
```

## Find (up to) four-gram phrases

```python
P3 = Phraser(Phrases(P2[P1[list(df['tokens'])]], threshold=10))
```

```python
for s in take(5, df['tokens']):
    print(' '.join(P3[P2[P1[s]]]))
    print()
```
